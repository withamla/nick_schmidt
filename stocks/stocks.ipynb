```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from twython import Twython
```

```python
apple = pd.read_csv('AAPL.csv', delimiter = ',')
google = pd.read_csv('GOOG.csv', delimiter = ',')
amazon = pd.read_csv('AMZN.csv', delimiter = ',')
```

```python
import matplotlib.dates as mdates #import date parsing function in matplotlib

google['Date'] = pd.to_datetime(google['Date']) #Convert column to date format and save to dataframe

google['High'] = pd.to_numeric(google['High']) #Convert to numeric and save to dataframe
google['Low'] = pd.to_numeric(google['Low']) #Convert to numeric and save to dataframe

amazon['Date'] = pd.to_datetime(amazon['Date']) #Convert column to date format and save to dataframe

amazon['High'] = pd.to_numeric(amazon['High']) #Convert to numeric and save to dataframe
amazon['Low'] = pd.to_numeric(amazon['Low']) #Convert to numeric and save to dataframe

google = google[(google['Date'] > '2008-1-1') & (google['Date'] <= '2008-12-31')]
amazon = amazon[(amazon['Date'] > '2008-1-1') & (amazon['Date'] <= '2008-12-31')]

# Plot
plt.figure(figsize=(30,10)) #set size
plt.plot(google['Date'], google['High'], color = 'green') #plot x date y high value for google
plt.plot(google['Date'], google['Low'], color = 'purple') #plot x date y low value for google
plt.plot(amazon['Date'], amazon['High'], color = 'blue') #plot x date y high value for amazon
plt.plot(amazon['Date'], amazon['Low'], color = 'red') #plot x date y low value for amazon
ax = plt.gca() #Gets polar max and min for axis
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1)) #Set interval to 4 months between points
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) #Format our date data as such
plt.gcf().autofmt_xdate() # Rotation for labels on x axis
plt.show()
```
